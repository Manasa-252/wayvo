# **K-Means Clustering**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Mall_Customers.csv')
df.head()
```

```python
df.isnull().any()
```

```python
X = df.iloc[:,[3,4]].values
print(X)
```

```python
from sklearn.cluster import KMeans
WCSS = []
for i in range(1,21):
  km = KMeans(n_clusters=i, init='k-means++', random_state=0)
  km.fit(X)
  WCSS.append(km.inertia_)
```

```python
plt.plot(range(1,21),WCSS)
plt.title('Elbow Method')
plt.xlabel('Cluster')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0)
y_kmeans = kmeans.fit_predict(X)
print(X,y_kmeans)
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0,1],c='r', label='c1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1,1],c='g', label='c2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2,1],c='b', label='c3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3,1],c='y', label='c4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4,1],c='m', label='c5')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='black', label='centroid')
plt.title('clusters of customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```

```python
kmeans.predict([[60,25]])
```

# **Hierarchical Clustering**

```python
X = df.iloc[:,[3,4]].values
print(X)
```

```python
import scipy.cluster.hierarchy as sc
dendrogram = sc.dendrogram(sc.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, linkage='ward')
y_hc = hc.fit_predict(X)
print(X,y_hc)
```

```python
plt.scatter(X[y_hc==0,0], X[y_hc ==0,1], c='r', label='c1')
plt.scatter(X[y_hc==1,0], X[y_hc ==1,1], c='g', label='c2')
plt.scatter(X[y_hc==2,0], X[y_hc ==2,1], c='b', label='c3')
plt.scatter(X[y_hc==3,0], X[y_hc ==3,1], c='y', label='c4')
plt.scatter(X[y_hc==4,0], X[y_hc ==4,1], c='m', label='c5')

plt.title('Customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
